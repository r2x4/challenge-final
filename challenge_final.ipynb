# **Eliminar Columnas**

```python
datos_ml = datos.copy()
columnas_a_eliminar = ['ID_Cliente', 'Cargos_Totales', 'Cargos_Diarios']
datos_ml = datos_ml.drop(columns=columnas_a_eliminar)
```

```python
# Cracion de las variables
y = datos_ml['Cancelación']
X = datos_ml.drop('Cancelación', axis=1)
```

```python
# Verificamos
print("Forma final de X:", X.shape)
print("Variables en X:", X.columns.tolist())
print("Variable objetivo y:", y.name)
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
# modelo
from xgboost import XGBClassifier

modelo = XGBClassifier(n_estimators=100, max_depth=4, learning_rate=0.1, random_state=42)
modelo.fit(X, y)
modelo.score(X, y)
```

```python
print (f'este modelo me da precicion de: {modelo.score(X, y):.2%}')
```

```python
from sklearn.model_selection import train_test_split
```

```python
# División adecuada de los datos
X,X_test,y,y_test = train_test_split(X, y, test_size=0.15, random_state=5, stratify=y)
X_train,X_val,y_train,y_val = train_test_split(X, y, random_state=5, stratify=y)
```

```python
modelo = XGBClassifier(max_depth=10)
modelo.fit(X_train, y_train)
print(f'La exactitud de Modelo entrenamiento es : {modelo.score(X_train,y_train):.2%}')
print(f'La exactitud de Modelo validacion es : {modelo.score(X_val,y_val):.2%}')
```

```python
# Suponiendo que y es tu variable objetivo original
print(y_train.value_counts(normalize=True))
print(y_val.value_counts(normalize=True))
```

```python
# SE MEJORA BALANCE

from xgboost import XGBClassifier

# Calcula el peso como proporción inversa
weight = (y_train == 0).sum() / (y_train == 1).sum()

modelo = XGBClassifier(
    max_depth=4,
    scale_pos_weight=weight,
    eval_metric='logloss',
    random_state=42
)

modelo.fit(X_train, y_train)

print(f'La exactitud de Modelo entrenamiento es : {modelo.score(X_train, y_train):.2%}')
print(f'La exactitud de Modelo validacion es : {modelo.score(X_val, y_val):.2%}')
```

```python
from sklearn.metrics import classification_report, confusion_matrix, f1_score

# Predicción
y_pred = modelo.predict(X_val)

# Matriz de Confusión
cm = confusion_matrix(y_val, y_pred)
print("=== Matriz de Confusión ===")
print(f"Verdaderos negativos (0,0): {cm[0][0]}")
print(f"Falsos positivos   (0,1): {cm[0][1]}")
print(f"Falsos negativos   (1,0): {cm[1][0]}")
print(f"Verdaderos positivos (1,1): {cm[1][1]}")
print()

# Reporte de Clasificación
print("=== Reporte de Clasificación ===")
print(classification_report(y_val, y_pred, digits=2))

# F1 Score
f1 = f1_score(y_val, y_pred)
print(f"=== F1 Score Global: {f1:.4f} ===")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Predicciones
y_pred = modelo.predict(X_val)

# Métricas de evaluación
exactitud = accuracy_score(y_val, y_pred)
precision = precision_score(y_val, y_pred)
recall = recall_score(y_val, y_pred)
f1 = f1_score(y_val, y_pred)

# Mostrar resultados
print(f"Exactitud (Accuracy): {exactitud:.2%}")
print(f"Precisión (Precision): {precision:.2%}")
print(f"Exhaustividad (Recall): {recall:.2%}")
print(f"F1 Score: {f1:.2%}")
```

```python
import numpy as np
from sklearn.metrics import precision_recall_curve

# Obtener probabilidades
y_prob = modelo.predict_proba(X_val)[:, 1]

# Evaluar distintos umbrales
precisions, recalls, thresholds = precision_recall_curve(y_val, y_prob)

# Graficar curva Precisión vs Recall
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(thresholds, precisions[:-1], label='Precisión')
plt.plot(thresholds, recalls[:-1], label='Recall')
plt.xlabel('Umbral')
plt.ylabel('Métrica')
plt.title('Precisión vs Recall según el umbral')
plt.legend()
plt.grid(True)
plt.show()
```

```python
umbral = 0.50  # puedes probar diferentes
y_pred_ajustado = (y_prob >= umbral).astype(int)

print("Nueva evaluación con umbral ajustado:")
print(f"Exactitud: {accuracy_score(y_val, y_pred_ajustado):.2%}")
print(f"Precisión: {precision_score(y_val, y_pred_ajustado):.2%}")
print(f"Recall: {recall_score(y_val, y_pred_ajustado):.2%}")
print(f"F1 Score: {f1_score(y_val, y_pred_ajustado):.2%}")
```
