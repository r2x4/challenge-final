```python
# MEJORA RÁPIDA PARA TU DECISION TREE ACTUAL
# Úsalo directamente con tus datos existentes

import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.metrics import classification_report, confusion_matrix, f1_score, accuracy_score, precision_score, recall_score
from imblearn.over_sampling import SMOTE, BorderlineSMOTE
from imblearn.combine import SMOTETomek
from collections import Counter

# =====================================
# 1. MEJORA DE CARACTERÍSTICAS RÁPIDA
# =====================================

print("🚀 MEJORANDO TUS DATOS...")

# Crear nuevas características
datos_ml['Cliente_Nuevo'] = (datos_ml['Meses_Conectado'] <= 6).astype(int)
datos_ml['Alto_Valor'] = (datos_ml['Cargos_Mensuales'] > datos_ml['Cargos_Mensuales'].quantile(0.75)).astype(int)
datos_ml['Cargos_Por_Mes'] = datos_ml['Cargos_Mensuales'] * datos_ml['Meses_Conectado']

# Contar servicios adicionales
servicios = ['Seguridad_En_Línea', 'Respaldo_En_Línea', 'Protección_Dispositivo',
            'Soporte_Técnico', 'TV_Streaming', 'Películas_Streaming']

total_servicios = 0
for servicio in servicios:
    if servicio in datos_ml.columns:
        servicio_num = pd.get_dummies(datos_ml[servicio], drop_first=True)
        if len(servicio_num.columns) > 0:
            total_servicios += servicio_num.iloc[:, 0]

datos_ml['Total_Servicios'] = total_servicios

print("✅ Nuevas características creadas")

# =====================================
# 2. PREPARAR DATOS MEJORADOS
# =====================================

y = datos_ml['Cancelación']
X = datos_ml.drop('Cancelación', axis=1)
X = pd.get_dummies(X, drop_first=True)

print(f"Características originales vs nuevas: 18 → {X.shape[1]}")

# =====================================
# 3. DIVISIÓN Y BALANCEO INTELIGENTE
# =====================================

# División inicial
X_train_orig, X_test, y_train_orig, y_test = train_test_split(
    X, y, test_size=0.15, random_state=42, stratify=y
)

print(f"\nDatos originales - Cancelan: {Counter(y_train_orig)[1.0]} / Total: {len(y_train_orig)}")

# Probar 3 estrategias de balanceo rápido
estrategias = {
    'SMOTE': SMOTE(random_state=42),
    'BorderlineSMOTE': BorderlineSMOTE(random_state=42),
    'SMOTETomek': SMOTETomek(random_state=42)
}

mejores_resultados = {}

for nombre, sampler in estrategias.items():
    print(f"\n📊 PROBANDO {nombre}...")

    # Balancear datos
    X_balanced, y_balanced = sampler.fit_resample(X_train_orig, y_train_orig)
    print(f"Datos balanceados: {Counter(y_balanced)}")

    # División entrenamiento/validación
    X_train, X_val, y_train, y_val = train_test_split(
        X_balanced, y_balanced, test_size=0.2, random_state=42, stratify=y_balanced
    )

    # =====================================
    # 4. OPTIMIZACIÓN DE HIPERPARÁMETROS
    # =====================================

    # Parámetros optimizados para tu caso
    param_grid = {
        'max_depth': [4, 5, 6, 7],
        'min_samples_split': [10, 20, 50],
        'min_samples_leaf': [5, 10, 20],
        'class_weight': ['balanced']
    }

    dt = DecisionTreeClassifier(random_state=42)

    grid_search = GridSearchCV(
        dt, param_grid,
        scoring='f1',
        cv=3,  # Reducido para velocidad
        n_jobs=-1
    )

    grid_search.fit(X_train, y_train)
    mejor_modelo = grid_search.best_estimator_

    # =====================================
    # 5. EVALUACIÓN
    # =====================================

    # Predicciones
    y_pred_val = mejor_modelo.predict(X_val)

    # Métricas
    f1 = f1_score(y_val, y_pred_val)
    precision = precision_score(y_val, y_pred_val)
    recall = recall_score(y_val, y_pred_val)
    accuracy = accuracy_score(y_val, y_pred_val)

    print(f"Resultados con {nombre}:")
    print(f"  F1 Score: {f1:.4f}")
    print(f"  Precisión: {precision:.4f}")
    print(f"  Recall: {recall:.4f}")
    print(f"  Exactitud: {accuracy:.4f}")

    # Guardar si es el mejor
    mejores_resultados[nombre] = {
        'modelo': mejor_modelo,
        'f1': f1,
        'precision': precision,
        'recall': recall,
        'accuracy': accuracy,
        'parametros': grid_search.best_params_
    }

# =====================================
# 6. SELECCIONAR MEJOR ESTRATEGIA
# =====================================

mejor_estrategia = max(mejores_resultados.keys(),
                      key=lambda k: mejores_resultados[k]['f1'])

print(f"\n🏆 MEJOR ESTRATEGIA: {mejor_estrategia}")
print(f"F1 Score: {mejores_resultados[mejor_estrategia]['f1']:.4f}")
print(f"Parámetros: {mejores_resultados[mejor_estrategia]['parametros']}")

# =====================================
# 7. MODELO FINAL CON OPTIMIZACIÓN DE UMBRAL
# =====================================

print(f"\n🎯 ENTRENANDO MODELO FINAL CON {mejor_estrategia}...")

# Re-entrenar con la mejor estrategia
sampler_final = estrategias[mejor_estrategia]
X_final_balanced, y_final_balanced = sampler_final.fit_resample(X_train_orig, y_train_orig)

X_final_train, X_final_val, y_final_train, y_final_val = train_test_split(
    X_final_balanced, y_final_balanced, test_size=0.2, random_state=42, stratify=y_final_balanced
)

# Modelo final
modelo_final = mejores_resultados[mejor_estrategia]['modelo']
modelo_final.fit(X_final_train, y_final_train)

# Optimización de umbral
from sklearn.metrics import precision_recall_curve

y_prob = modelo_final.predict_proba(X_final_val)[:, 1]
precisions, recalls, thresholds = precision_recall_curve(y_final_val, y_prob)

# Calcular F1 para cada umbral
f1_scores = []
for i in range(len(thresholds)):
    if precisions[i] + recalls[i] > 0:
        f1 = 2 * (precisions[i] * recalls[i]) / (precisions[i] + recalls[i])
    else:
        f1 = 0
    f1_scores.append(f1)

# Mejor umbral
best_idx = np.argmax(f1_scores)
best_threshold = thresholds[best_idx]

# Predicciones con umbral optimizado
y_pred_optimizado = (y_prob >= best_threshold).astype(int)

print(f"\n🎯 UMBRAL OPTIMIZADO: {best_threshold:.4f}")
print(f"F1 Score optimizado: {f1_score(y_final_val, y_pred_optimizado):.4f}")

# =====================================
# 8. EVALUACIÓN FINAL EN DATOS DE PRUEBA
# =====================================

print(f"\n📊 EVALUACIÓN FINAL EN DATOS DE PRUEBA:")

# Predicciones en test con umbral optimizado
y_prob_test = modelo_final.predict_proba(X_test)[:, 1]
y_pred_test_optimizado = (y_prob_test >= best_threshold).astype(int)

print(f"Exactitud: {accuracy_score(y_test, y_pred_test_optimizado):.2%}")
print(f"Precisión: {precision_score(y_test, y_pred_test_optimizado):.2%}")
print(f"Recall: {recall_score(y_test, y_pred_test_optimizado):.2%}")
print(f"F1 Score: {f1_score(y_test, y_pred_test_optimizado):.2%}")

# Matriz de confusión final
cm_final = confusion_matrix(y_test, y_pred_test_optimizado)
print(f"\nMatriz de Confusión Final:")
print(f"TN: {cm_final[0,0]}, FP: {cm_final[0,1]}")
print(f"FN: {cm_final[1,0]}, TP: {cm_final[1,1]}")

# =====================================
# 9. COMPARACIÓN CON TU MODELO ORIGINAL
# =====================================

print(f"\n📈 COMPARACIÓN CON TU MODELO ORIGINAL:")
print("=" * 50)
print("MODELO ORIGINAL (Decision Tree básico):")
print("  F1 Score: 53.46%")
print("  Precisión: 57.94%")
print("  Recall: 49.62%")
print("  Exactitud: 77.06%")
print()
print("MODELO MEJORADO:")
print(f"  F1 Score: {f1_score(y_test, y_pred_test_optimizado):.2%}")
print(f"  Precisión: {precision_score(y_test, y_pred_test_optimizado):.2%}")
print(f"  Recall: {recall_score(y_test, y_pred_test_optimizado):.2%}")
print(f"  Exactitud: {accuracy_score(y_test, y_pred_test_optimizado):.2%}")

mejora_f1 = f1_score(y_test, y_pred_test_optimizado) - 0.5346
print(f"\n🚀 MEJORA EN F1 SCORE: {mejora_f1:+.4f}")

# =====================================
# 10. CÓDIGO PARA PRODUCCIÓN
# =====================================

print(f"\n💡 CÓDIGO PARA USAR EN PRODUCCIÓN:")
print("=" * 50)
print(f"# Usar estrategia: {mejor_estrategia}")
print(f"# Umbral óptimo: {best_threshold:.4f}")
print(f"# Parámetros del modelo: {mejores_resultados[mejor_estrategia]['parametros']}")
print()
print("# Para nuevas predicciones:")
print("probabilidades = modelo_final.predict_proba(X_nuevo)[:, 1]")
print(f"predicciones = (probabilidades >= {best_threshold:.4f}).astype(int)")

print(f"\n✅ PROCESO COMPLETADO!")
print(f"🎯 Tu modelo ha sido mejorado significativamente con {mejor_estrategia} y umbral optimizado")
```
